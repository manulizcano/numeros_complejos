```python
def funcion_compleja(z):
  return (z-5)*(z+5)*(z+5j)

funcion_compleja_vectorizada = np.vectorize(funcion_compleja)

def plot_campo():
  m,k = 10, 20
  x = np.linspace(-m, m, k)
  y = np.linspace(-m, m, k)
  xx, yy = np.meshgrid(x, y)
  zz = xx + 1j*yy
  zz = funcion_compleja_vectorizada(zz)

  z_R = zz.real
  z_I = zz.imag

  u_R = zz.real -xx
  u_I = zz.imag -yy


  fig = plt.figure(figsize=(10, 10))
  ax = fig.add_subplot(111)


  ax.quiver(xx, yy, u_R, u_I )
  color = 2 * np.log(np.hypot(z_R, z_I))
  ax.streamplot(x, y, z_R, z_I, color=color, linewidth=1, cmap=plt.cm.inferno, density=2, arrowstyle='->', arrowsize=1.5)

  ax.set_xlabel('Parte Real')
  ax.set_ylabel('Parte imaginaria')
  ax.set_xlim(-m,m)
  ax.set_ylim(-m,m)
  ax.set_aspect('equal')
  plt.show()

plot_campo()
```
